## 如何读取、绘制声音

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from scipy.io import wavfile
```

## 使用wavfile中的input_read读取音频

```python
freq, sound= wavfile.read('201.wav')
```

## 输出信号的相关参数

```python
print("声音维度：",sound.shape)
print("声音的数据类型：",sound.dtype)
print("声音的时长：{}秒".format(sound.shape[0]/freq))
```

## 归一化声音数据

```python
#这里我不知道为什么声音的归一化是除以2.**15，也就是15970
sound= sound/2.**15
```

## 绘制波形图

```python
x= np.arange(0,len(sound),1)/freq*1000
```

```python
plt.plot(x,sound)
```

## 频域分析

```python
#对声音做傅里叶变换
transformed_signal= np.fft.fft(sound)
half_length= int(len(sound)/2)
transformed_signal= abs(transformed_signal[0:half_length])
transformed_signal= transformed_signal/len(sound)*2
#归一化
transformed_signal= transformed_signal/len(sound)
#将幅度转化成功率
transformed_signal= transformed_signal**2
power= 10*np.log10(transformed_signal)
x_fft= np.arange(0,half_length,1)*(freq/len(sound))/1000.0
plt.plot(x_fft,power)
```

## 用python进行傅里叶变换

```python
#N表示采样的数量
N=600
#Fs表示采样频率
Fs=800
T= 1.0/Fs
#x表示时域样本点之间的间隔
x= np.linspace(0,N/Fs,N)
#signal_freq表示信号频率
signal_freq=10
#y表示信号
y= np.sin(2*np.pi*signal_freq*x)+np.cos(2*np.pi*50*x)
#对y进行FFT
y_fft= abs(np.fft.fft(y))
#设置频率间隔
k= np.arange(N)
sample_time= N/Fs
x_fft= k/sample_time
```

```python
plt.plot(x,y)
```

```python
plt.plot(x_fft, y_fft)
```

```python
plt.plot(x_fft[:N//2],y_fft[:N//2])
```

## 从上面的分析我们可以看出声音其实就是多种频率的正弦波的叠加，也就是说我们用叠加的正弦波也可以创造声音

```python
#创造声音出来
from scipy.io.wavfile import write
output_file= "output_generated.wav"
#生成一个长3s，采样频率为44100HZ，音频频率为587HZ的音频
duration= 3
sampling_fre= 44100
signal_fre= 587
min_val= -2*np.pi
max_val= -2*np.pi
t= np.linspace(min_val, max_val, duration*sampling_fre)
signal= np.sin(2*np.pi*signal_fre*t)
#增加噪声
noise= 0.4*np.random.rand(duration*sampling_fre)
signal+=noise
#转化成int16
scaling_factor= 2**15-1
signal_normalized= signal/np.max(np.abs(signal))
signal_scaled= np.int16(signal_normalized*scaling_factor)
write(output_file,sampling_fre,signal_scaled)
```

```python
#绘制出上面的信号
audio= signal[0:100]
x_time= np.arange(0,len(audio),1)/sampling_fre*1000
plt.plot(x_time, audio)
```

## 我们刚刚通过正弦波和一些无聊的噪声成功的创造了更无聊的噪声，下面来点硬的，创造音乐

```python
# 假如我们直到每个调调的频率，合起来不就成了音乐？
def synthesizer(freq, duration,amp=1.0,sampling_freq=44100):
    #时间轴
    t= np.linspace(0,duration, duration*sampling_freq)
    audio= amp*np.sin(2*np.pi*freq*t)
    return audio.astype(np.int16)
import json
#加载json格式的音阶
tone_map_file="tone_freq_map.json"
with open(tone_map_file,"r") as f:
    tone_freq_map= json.loads(f.read())
```

```python
import os
import argparse
import numpy as np
from scipy.io import wavfile
from hmmlearn import hmm 
from python_speech_features import mfcc
#创建一个HMM类
class HMMTrainer(object):
    def __init__(self, model_name="GaussianHMM",n_components=4,cov_type="diag",n_iter=1000):
        self.model_name=model_name
        self.n_components= n_components
        self.cov_type= cov_type
        self.n_iter= n_iter
        self.models=[]
        if self.model_name=="GaussianHMM":
            self.model= hmm.GaussianHMM(n_components=self.n_components,covariance_type=self.cov_type
            ,n_iter=self.n_iter)
        else:
            raise TypeError("Invalid model type")
    def train(self,X):
        np.seterr(all="ignore")#设置发生异常时不采取任何操作
        self.models.append(self.model.fit(X))
    def get_score(self,input_data):
        return self.model.score(input_data)


input_folder= r'C:\Users\Administrator\Desktop\资料\我的U盘\资料\机器学习\python机器学习经典实例\Chapter07\train'
hmm_models=[]
#解析包含所有数据库音频文件的输入路径
for dirname in os.listdir(input_folder):
    #获取子文件夹名称
    subfolder= os.path.join(input_folder,dirname)
    #判断是不是目录，如果不是就找下一个
    if not os.path.isdir(subfolder):
        continue
    #提取标签,rfind("/")是获取"/"最后一次出现的位置
    label= subfolder[subfolder.rfind("\\")+1:]
    #初始化用于训练的变量
    X= np.array([])
    y_words=[]
    #迭代每个子文件夹中的音频文件，x.endswith(".wav")用于判断是否以".wav"结束
    for filename in [x for x in os.listdir(subfolder) if x.endswith(".wav")][:-1]:
        #读取每个音频文件
        filepath= os.path.join(subfolder, filename)
        sampling_freq, audio= wavfile.read(filepath)
        #提取MFCC特征
        mfcc_features= mfcc(audio,sampling_freq)
        #将MFCC特征添加到变量X 
        if len(X)==0:
            X= mfcc_features
        else:
            X= np.append(X,mfcc_features,axis=0)
        #添加标签信息
        y_words.append(label)
    #训练并保存HMM模型
    hmm_trainer= HMMTrainer()
    hmm_trainer.train(X)
    hmm_models.append((hmm_trainer,label))
    hmm.trainer=None 
#测试文件
input_files_load=r'C:\Users\Administrator\Desktop\资料\我的U盘\资料\机器学习\python机器学习经典实例\Chapter07\test'
test_files= os.listdir(input_files_load)
for test_file in test_files:
    sampling_freq,audio= wavfile.read(os.path.join(input_files_load,test_file))
    mfcc_features= mfcc(audio,sampling_freq)
    #定义两个变量用于存放最大分数值和输出标记
    max_score=float("-inf") 
    output_label= None 
    for item in hmm_models:
        hmm_model, label= item
        score= hmm_model.get_score(mfcc_features)
        if score>max_score:
            max_score= score 
            output_label= label 
    #打印真实的、预测的标签
    print("True Label is:",test_file[:test_file.rfind("15")])
    print("Predicted Label is:",output_label)
    print()
```
