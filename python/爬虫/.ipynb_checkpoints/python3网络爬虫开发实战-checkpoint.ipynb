## urllib库

```python
#基本使用方法
import urllib.request
response= urllib.request.urlopen("https://python.org")
print(response.read().decode("utf-8"))
```

response是一个HTTPResponse类型的对象，主要包括read()、readinfo()、getheader(name)、getheaders()、fileno()方法以及status等属性

```python
print(response.status)
print(response.getheaders())
```

### urlopen的参数

```python
#timeout参数，用于设置请求时间，超时返回异常
import urllib.request
url= "https://python.org"
response= urllib.request.urlopen(timeout=0.1,url=url)
print(response.read().decode("utf-8"))
```

```python
#捕获超时错误
import urllib.error
import urllib.request
import socket
try:
    url= "https://python.org"
    response= urllib.request.urlopen(url= url,timeout=0.1)
except urllib.error.URLError as e:
    if isinstance(e.reason,socket.timeout):
        print("TIME OUT")
```

### Request方法

```python
#通过Request灵活的构造参数
import urllib.request
url= "https://python.org"
headers={"User-Agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/72.0.3626.119 Safari/537.36"}
request= urllib.request.Request(url,headers=headers)
response= urllib.request.urlopen(request)
print(response.read().decode("utf-8"))
```

### 异常处理

```python
#访问到一个不存在的网站，捕获异常
import urllib.request
from urllib import error
try:
    url= "http://www.cuiqingcai.com/index.htm"
    response= urllib.request.urlopen(url=url)
except error.URLError as e:
    print(e.reason)
```

## 正则表达式

正则表达式主要用于HTML中信息的提取

常用的一些匹配规则  
\w：匹配字母数字和下划线  
\W：匹配非字母数字下划线  
\s：匹配任意空白字符，等价于[\t\n\r\f] 
\S：匹配任意非空字符
\d：匹配任意数字  
\D：匹配任意非数字  
^：匹配开头  
$：匹配结尾  
.：匹配一个非换行符  
*：匹配0个或多个字符  
+：至少匹配一个字符  
？：最多匹配一个字符  

### match()匹配方法

match()尝试从字符串起始位置匹配正则表达式，如果匹配就返回匹配结果，如果不匹配就返回None

```python
import re
content= "Hello 123 4567 World_This is a Regex Demo"
patent= "Hello\s\d{3}\s\d{4}\s\w{10}"
result= re.match(patent, content)
print(result)
```

```python
print(result.group())
```

如果你想将字符串其中的一部分提取出来，可以使用()组实现

```python
content="Hello 1234567 World_This is a Regex Demo"
patent= "^Hello\s(\d+)\sWorld"
result= re.match(patent, content)
print(result.group())
print(result.groups())
print(result.group(1))
```

### 贪婪匹配和非贪婪匹配

.* 尽可能匹配多个字符，属于贪婪匹配  
.\*? 尽可能少的匹配字符，属于非贪婪匹配  
正则表达式中尽可能多用非贪婪匹配

```python
content="Hello 1234567 World_This is a Regex Demo"
patent= "He.*(\d+).*Demo$"
result= re.match(patent, content)
print(result.group(1))
```

本来想匹配内容中的数字，结果只匹配到了一个数字

```python
content="Hello 1234567 World_This is a Regex Demo"
patent= "He.*?(\d+).*?Demo$"
result= re.match(patent, content)
print(result.group(1))
```

### 常用修饰符

re.I：对大小写不敏感  
re.M：多行匹配  
re.S：匹配包括换行符在内的所有字符

### 转义匹配

对于一些特殊字符的匹配，通常加上转移字符，是指成为普通字符

```python
import re
content="(百度)www.baidu.com"
result= re.match("\(百度\)www\.baidu\.com",content)
print(result.group(0))
```

### search()方法
和match()方法的区别在于search()不是从开头开始匹配的，只要有就行

```python
import re
content="Extra strings Hello 1234567 World_This is a Regex Demo"
pattern= "Hello.*?(\d+).*?Demo$"
result= re.search(pattern, content)
print(result.group(1))
```

### findall()方法
和search()方法不同在于，findall()搜寻整个字符串，找到所有和匹配项

### sub()方法
sub()方法用于修改文本

```python
import re
content="54Ky215245YFEJO"
pattern="\d+"
result= re.sub(pattern,"",content)
print(result)
```

### compile()方法
将正则字符串编译成表达式对象，相当于给正则表达式做了封装，方便以后复用

## 实战：抓取猫眼电影排行

任务：利用requests库和正则表达式抓取猫眼电影TOP 100的相关内容(100个电影的名称、时间、评分、图片)  
抓取的网址为：[TOP 100榜](https://maoyan.com/board/4)

```python
import requests
def get_one_page(url):
    #请求头
    headers={"User-Agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/72.0.3626.119 Safari/537.36"}
    response= requests.get(url, headers= headers)
    if response.status_code==200:
        return response.text
    return None
def main():
    url="https://maoyan.com/board/4"
    html= get_one_page(url)
    print(html)
```

```python
html= get_one_page("https://maoyan.com/board/4")
```

```python
print(html)
```
